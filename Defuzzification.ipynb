```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
x = np.arange(0, 5.05, 0.1)

mfx = fuzz.trapmf(x, [2, 2.5, 3, 4.5])
```

```python
modes = ['centroid', 'bisector', 'mom', 'som', 'lom']
xvals = [fuzz.defuzz(x, mfx, mode) for mode in modes]

colors = ['r', 'b', 'g', 'c', 'm']

ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, mfx, 'k')

for xv, y, mode, color in zip(xvals, ymax, modes, colors):
    plt.vlines(xv, 0, y, label=mode, color=color)
    
plt.xlabel('Universe Variables')
plt.ylabel('Fuzzy Membership')

plt.ylim(-0.1, 1.1)
plt.legend(loc=2)
plt.show()
```
